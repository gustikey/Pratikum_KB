```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('Maths.csv')
```

```python
df
```

```python
x = list(df["sex"])
y = list(df["age"])

plt.figure(figsize=(5,6))
plt.bar(x, y, color='red')
plt.title("Perbandingan Gender di sekolah")
plt.xlabel("Jenis Kelamin")
plt.ylabel("Umur")
```

Kesimpulan

Dari Grafik diatas bisa kita lihat perbandingan gender sekolah berdasarkan 2 kategori,  Gender Disekolah dengan performa tertinggi  adalah Gender M, Dan gender disekolah dengan performa terendah adalah Gender F.

```python
df["sex"].value_counts()
```

```python
data_pie = df["sex"].value_counts().rename_axis("Jenis Kelamin").reset_index(name="jumlah")
```

```python
data_pie.head(10)
```

```python
plt.pie(data_pie["jumlah"],labels=data_pie['Jenis Kelamin'],startangle=90,autopct='%.1f%%' ,explode=(0,0))
plt.title('Persentase Perbandingan Jenis Kelamin')
plt.show()
```

Kesimpulan

dari grafik diatas kita bisa melihar presentase skor gender disekolah. dan peforma gender disekolah yang tertinggi adalah gender F yaitu 52,7%, sedangkan  peforma gender disekolah yang terrendah adalah gender M yaitu 47,3%.

```python
sns.scatterplot(data=df,x ="sex",y= 'age')
plt.title('Perbandingan Jenis Kelamin dan Umur')
plt.xlabel('Umur')
plt.ylabel('Jenis Kelamin')
plt.show()
```

Kesimpulan
