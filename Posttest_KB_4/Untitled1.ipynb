# Data Cleaning

```python
from sklearn.impute import SimpleImputer
```

```python
df.isna().sum()
```

# Mengubah nilai null yang punya tipe data desimal

```python
imputer = SimpleImputer(strategy = 'mean')
df["height_cm"] = imputer.fit_transform(df[["age"]])
print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```

# Mengubah nilai null yang punya tipe data kategorikal

```python
imputer = SimpleImputer(strategy = 'most_frequent')
df["weight_kg"] = imputer.fit_transform(df[["weight_kg"]])
print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```

# Mengubah nilai null yang punya tipe data bulat

```python
imputer = SimpleImputer(strategy = 'most_frequent')
df["age"] = imputer.fit_transform(df[["age"]])
print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```
